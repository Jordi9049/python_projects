```python
#First I will set the independent variables
X = df[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'smoking_status_formerly smoked', 'smoking_status_never smoked', 'smoking_status_smokes']]

#Now I will set the dependent variable
y = df['Stroke']
```

## Splitting the data into Train and Test

```python
# I will now divide the data into train and test spilt
# I have split the data into 80/20 as this would be considered standard when working with a large dataset and I chose 11 as the random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=11)
```

```python
#Now I will scale the data
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

## First machine learning algorithm - Logistic Regression

```python
classifier = LogisticRegression(random_state=11, class_weight='balanced')
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
#Now I can make the predictions
y_pred=classifier.predict(X_test)
y_pred[:200]
```

## Evaluating the model

```python
#I will now create the Confusion matrix and Classification report to check the accuracy of the model

print('Confusion Matrix \n',confusion_matrix(y_test,y_pred))
print('\n')
print('Classification Report\n', classification_report(y_test, y_pred, zero_division=0))
print('\n')
print('Accuracy',accuracy_score(y_test,y_pred))
```

```python
#I will now create the Confusion matrix and Classification report to check the accuracy of the model

print('Confusion Matrix \n',confusion_matrix(y_test,y_pred))
print('\n')
print('Classification Report\n',classification_report(y_test,y_pred))
print('\n')
print('Accuracy',accuracy_score(y_test,y_pred))
```
