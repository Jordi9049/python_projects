### ***Logistic Regression***

Logistic regression is a statistical model that in its basic form uses a logistic function to model a binary dependent variable. It is simple, easy to implement, and provides good results for binary classification problems.

- For the HR retention problem, where the outcome is binary (employee stays or leaves), logistic regression provides a solid baseline for understanding the influence of various features on employee churn.

- It's interpretative and less complex compared to more sophisticated models, making it easier to explain the results.




### ***K-Nearest Neighbors (KNN)***
KNN is a simple, instance-based learning algorithm where the function is only approximated locally, and all computation is deferred until function evaluation. It is highly versatile and easy to understand.

- In scenarios where the pattern of employee turnover is not well-defined by boundaries, KNN can predict the outcome for new data points based on a similarity score from the nearest data points.

- Useful for its simplicity and effectiveness in classification problems on smaller datasets or where decision boundaries are very irregular.

### ***Random Forest***

A type of ensemble learning method, Random Forest builds multiple decision trees and merges them together to get a more accurate and stable prediction. It is known for its robustness and good accuracy.

- Given the HR data might include interactions and nonlinearities that single decision trees miss, Random Forest can capture these effectively, providing better accuracy without the risk of overfitting.

- Offers higher accuracy through ensemble learning and handles both numerical and categorical data well, making it ideal for complex datasets.

### ***Why Other Algorithms Were Not Chosen***

- `Decision Trees`: While intuitive, decision trees often suffer from overfitting, especially with datasets having many features. Random Forest mitigates this by averaging multiple trees, thus providing better generalization.

- `Linear Regression`: Typically used for predicting continuous dependent variables, linear regression was not chosen because the outcome of the HR retention is binary. Linear regression would not naturally accommodate the categorical nature of the response variable without modifications, such as logistic regression, which models the probability of class membership.

## Converting Categorical data into Numerical

I need to convert our categorical data to numerical values. Python cannot work with nominal/catgorical type data.

This is integral for visualisations as it requires numerical data
There are a few approaches we can take

* `Mapping`
* `Labelling`
* `One-hot encoding `

As I want to visualise the data, these conversions were left until after the data was cleaned up. 

Labels and text data can be useful in charts, but not so useful for machine learning


```python
from sklearn.preprocessing import LabelEncoder #Importing a new library, label encoding for lav

# Set display options to show more columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', 1000)

# Copy the data_deduped DataFrame to a new one for conversion. Tracks progress
data_converted = data_deduped.copy()

# Applying mapping for 'Work_accident' and 'left'
mapping = {'No': 0, 'Yes': 1}
data_converted['Work_accident'] = data_converted['Work_accident'].map(mapping)
data_converted['left'] = data_converted['left'].map(mapping)

# Label encoding for 'promotion_last_5years'
label_encoder = LabelEncoder()
data_converted['promotion_last_5years'] = label_encoder.fit_transform(data_converted['promotion_last_5years'])

# One-Hot Encoding for 'Department' since there are more than three unqiue options
data_converted = pd.get_dummies(data_converted, columns=['Department'], drop_first=False)

# Manual One-Hot Encoding for 'salary' to keep the order low, medium, high
salary_order = ['low', 'medium', 'high']
for category in salary_order:
    data_converted[f'salary_{category}'] = (data_converted['salary'] == category).astype(int)

# Drop the original 'salary' column
data_converted.drop('salary', axis=1, inplace=True)

# Convert the one-hot encoded columns to integer type to ensure they are 0/1
one_hot_columns = [col for col in data_converted if col.startswith('Department_')]
data_converted[one_hot_columns] = data_converted[one_hot_columns].astype(int)

# Verify the changes
data_converted.head()
```

# **Models**

### ***Understanding Model Evaluation Metrics***

After training our models, we evaluate them using several metrics to understand their performance:

- **Confusion Matrix**: This matrix helps visualize the performance of a model. Each row of the matrix represents the instances in an actual class, while each column represents the instances in a predicted class. It’s particularly useful for seeing how many instances are correctly or incorrectly classified in each class.

    - **True Negatives (TN)**: The top-left quadrant shows the number of predictions where the model correctly predicted the employees who stayed.
    - **False Positives (FP)**: The top-right quadrant indicates the number of employees who stayed but were incorrectly predicted as having left.
    - **False Negatives (FN)**: The bottom-left quadrant represents the employees who left but were incorrectly predicted as having stayed.
    - **True Positives (TP)**: The bottom-right quadrant shows the correct predictions of employees who left.

- **Classification Report**

    - **Accuracy**: This is the simplest metric, calculated as the proportion of correctly predicted observations to the total observations. It gives us a quick indication of how well the model is performing overall.
    - **Precision and Recall**: These metrics are crucial for datasets where some classes are more important to correctly predict than others. Precision tells us out of all the predicted positive classes, how many are actually positive. Recall tells us out of all the actual positive classes, how many we predicted correctly.
    - **F1-Score**: This metric combines precision and recall into a single metric by taking their harmonic mean. It is useful when you need a balance between precision and recall.

#### *Why These Metrics Matter*
Understanding these metrics helps in assessing the strengths and weaknesses of the models, particularly in how they handle different types of errors (like false positives vs. false negatives).

Choosing `n_neighbors=5` is a common starting point that often performs well in a variety of scenarios without specific tuning.
